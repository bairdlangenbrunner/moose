```python
import numpy
from netCDF4 import Dataset
import matplotlib.pyplot as mp
import matplotlib.colors as mc
import matplotlib.cm as cm
import mpl_toolkits.mplot3d
import matplotlib
import scipy.ndimage
import datetime

import itertools
import random
import numpy.random
import scipy.stats
import os

import matplotlib.patches

mp.rcParams.update({'mathtext.default': 'regular'})

from mpl_toolkits import basemap
import mpl_toolkits.axes_grid1

degree_sign = u'\u00B0'
% matplotlib inline
```

```python
fig = mp.figure(figsize=(5,5))
ax = fig.add_subplot(111)

a = numpy.random.rand(1000)
b = numpy.random.rand(1000)

x_error = 0.41
y_error = 0.24

ellipse_plot = matplotlib.patches.Ellipse(xy=[0,0], width=2*x_error, height=2*y_error, angle=0, facecolor='0.5', edgecolor='0.5', zorder=1)

ax.scatter(a,b, zorder=2)

ax.tick_params(labelsize=12)
ax.grid()
ax.set_axisbelow(True)
ax.add_artist(ellipse_plot)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
```
